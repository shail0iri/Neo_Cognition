```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === LOAD CEW CSV ===
df = pd.read_csv(r"C:\Users\Shail\Downloads\neo_cognition\outputs\cew_processed_dataset.csv")

print("\n================ CEW DATASET ANALYSIS ================\n")

# --- BASIC INFO ---
print(f"Total samples: {len(df):,}")
print("\nClass Balance:")
print(df['eye_state'].value_counts())

print("\nSplit Distribution:")
print(df['split'].value_counts())

print("\nEye Side Distribution:")
print(df['eye_side'].value_counts())

print("\nAugmentation Distribution:")
print(df['augmented'].value_counts())


# =============== PLOTS ===============

plt.figure(figsize=(10,5))
df['eye_state'].value_counts().plot(kind='bar', color=['green','red'])
plt.title("Class Distribution (Open vs Closed)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.grid(axis='y', alpha=0.3)
plt.show()


plt.figure(figsize=(10,5))
df['split'].value_counts().plot(kind='bar', color=['blue','orange','purple'])
plt.title("Train / Val / Test Split Distribution")
plt.xlabel("Split")
plt.ylabel("Count")
plt.grid(axis='y', alpha=0.3)
plt.show()


plt.figure(figsize=(10,5))
df['eye_side'].value_counts().plot(kind='bar', color=['cyan','magenta'])
plt.title("Left vs Right Eye Distribution")
plt.xlabel("Eye Side")
plt.ylabel("Count")
plt.grid(axis='y', alpha=0.3)
plt.show()


plt.figure(figsize=(6,6))
df['augmented'].value_counts().plot(kind='pie', autopct='%1.1f%%', labels=['Original','Augmented'], colors=['gold','lightblue'])
plt.title("Original vs Augmented Samples")
plt.ylabel("")
plt.show()


# === CORRELATION HEATMAP ===
plt.figure(figsize=(6,5))
sns.heatmap(df[['label','augmented']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

print("\n CEW ANALYSIS COMPLETE!\n")
```
